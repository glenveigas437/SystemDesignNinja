Requirements
- 2 Players
- Players take turns in an alternate manner


Unclear Requirements
- Game Terminates:
    - Either player wins
    - Draw

Rules
- Player 1 puts X
- Player 2 puts O
- Game Termination: 
    - Draw: Board is full
    - Vertical/Horizontal/Diagonal elements are same 

Nouns
- Players, Game, Board, Symbol, Turn

```python
from abc import ABC, abstractmethod
import random as rd
```

```python
class Board:
    def __init__(self, size):
        self.size = size
        self.emptyCells=0
        self.board = [['_' for i in range(self.size)]for j in range(self.size)]
        self.refBoard = [['_' for i in range(self.size)]for j in range(self.size)]
        for i in range(len(self.refBoard)):
            for j in range(len(self.refBoard[0])):
                self.refBoard[i][j]=(i,j)
        
        
    def referenceBoard(self):
        for i in range(len(self.refBoard)):
            for j in range(len(self.refBoard[0])):
                print(self.refBoard[i][j], end=" ")

            print()
    
    def isBoardFull(self):
        return self.emptyCells == self.size * self.size
    
    def hasWinningLine(self):
        return True if self.__vertical() or self.__horizontal() or self.__diagonal() else False
    
    def __vertical(self):
        for col in range(self.size):
            currentSymbol = self.board[0][col]
            if currentSymbol == '_':
                continue
            count = 0
            row = 0
            while(row<self.size):
                if(self.board[row][col]!=currentSymbol):
                    break
                count+=1
                row+=1
            
            if(count==self.size):
                return True
        
        return False
    
    def __horizontal(self):
        for row in range(self.size):
            currentSymbol = self.board[row][0]
            if currentSymbol == '_':
                continue
            count = 0
            col = 0
            while(col<self.size):
                if(self.board[row][col]!=currentSymbol):
                    break
                count+=1
                col+=1
            
            if(count==self.size):
                return True
        
        return False
    
    def __diagonal(self):
        
        row=0
        col=0
        
        currentSymbol = self.board[row][col]
        if currentSymbol == '_':
            return False
        count=0
        while(row<self.size and col<self.size):
            if(self.board[row][col]==currentSymbol):
                count+=1
            else:
                break
            row+=1
            col+=1

        if(count==self.size):
            return True
        
        
        row = 0
        col = self.size-1
        
        currentSymbol = self.board[row][col]
        if currentSymbol == '_':
            return False
        count=0
        while(row<self.size and col<=0):
            if(self.board[row][col]==currentSymbol):
                count+=1
            else:
                return False
            row+=1
            col-=1

        if(count==self.size):
            return True  
    
    def makeMove(self, row, col, currentPlayer):
        self.board[row][col]=currentPlayer.symbol
        self.emptyCells+=1
        self.refBoard[row][col]=currentPlayer.symbol
                        
    def displayBoard(self):
        for row in range(self.size):
            for col in range(self.size):
                print(self.board[row][col], end=" ")

            print()            
```

```python
class Player(ABC):
    def __init__(self, name, symbol):
        self.name = name
        self.symbol = symbol
        self.type = None
    
    def selectSlot(self):
        pass
```

```python
class HumanPlayer(Player):
    def selectSlot(self, board):
        self.row, self.col = map(int, input(f"{self.name} please enter the co-ordinates of an empty slot: ").split())
        placed = False
        while(not placed):
            if(board.board[self.row][self.col]=='_'):                
                placed = True
            else:
                self.row, self.col = map(int, input(f"{self.name}, Invalid move! please enter the co-ordinates of an empty slot: ").split())
        return self.row, self.col
```

```python
class AI(Player):
    def selectSlot(self, board):
        self.row = rd.randint(0, board.size-1)
        self.col = rd.randint(0, board.size-1)
        placed = False
        while(not placed):
            if(board.board[self.row][self.col]=='_'):                
                placed = True
            else:
                self.row = rd.randint(0, board.size-1)
                self.col = rd.randint(0, board.size-1)
        print(f"AI Has placed it at {self.row, self.col}")
        return self.row, self.col
```

```python
class Game:
    def playGame(self):
        size = int(input("Enter Board Size: "))
        board = Board(size)
        player1Name, player1Symbol = map(str, input("Enter your name and Symbol: ").split())
        player1 = HumanPlayer(player1Name, player1Symbol)
        opponentType = int(input("1. AI, 2. Human"))
        
        
        if player1Symbol == 'X':
            player2Symbol = 'O'
        else:
            player2Symbol = 'X'
        if opponentType == 1:
            player2 = AI('CPU', player2Symbol)
        else:
            player2Name = input("Enter Your name: ")
            player2 = HumanPlayer(player2Name, player2Symbol)
        
        if player1.symbol == 'X':
            currentPlayer = player1
        elif player2.symbol == 'X':
            currentPlayer = player2
        
        while(not board.isBoardFull()):
            board.referenceBoard()
            board.displayBoard()
            row, col = currentPlayer.selectSlot(board)
            board.makeMove(row, col, currentPlayer)
            if board.hasWinningLine():
                board.displayBoard()
                print(f"{currentPlayer.name} has won the game")
                return 0
            if currentPlayer == player1:
                currentPlayer = player2
            elif currentPlayer == player2:
                currentPlayer = player1
            
            print("")
        
        board.displayBoard()
        return "Game has been ended in a draw"
```

```python
class PlayGame:
    def start(self):
        Game().playGame()
```

```python
PlayGame().start()
```
